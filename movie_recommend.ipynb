```python
#genres
#id
#keywords
#title
#overview
#cast
#crew

dt=dt[['genres','id','keywords','title','overview','cast','crew']]
```

```python
dt.isnull().sum()
dt.dropna(inplace=True)
dt.isnull().sum()
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
        
```

```python
def convert3(obj):
    L=[]
    counter =0
    for i in ast.literal_eval(obj):
         if counter != 3:
            L.append(i['name'])
            counter +=1
         else:
           break
    return L
```

```python
dt['crew']=dt['crew'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
          L.append(i['name'])
        break
    return L
```

```python
dt['overview'] = dt['overview'].apply(lambda x: [x])
dt['genres'] = dt['genres'].apply(lambda x: [x])
dt['keywords'] = dt['keywords'].apply(lambda x: [x])
dt['cast'] = dt['cast'].apply(lambda x: [x])
```

```python
dt['tags']=dt['genres']+dt['keywords']+dt['overview']+dt['cast']+dt['crew']
```

```python
dt
```

```python
new_df=dt[['id','title','tags']]
```

```python
new_df
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in text:
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = dt['tags'].apply(stem)
```

```python
new_df['tags'] = dt['tags'].astype(str)

```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()

similarity = cosine_similarity(vectors)
```

```python
cv.get_feature_names_out()
```

```python
similarity[1]
```

```python
def recommender(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)

        
```

```python
recommender("Avatar")
```
